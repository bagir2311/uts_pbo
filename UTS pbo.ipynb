# Implementasi Pewarisan Multilevel Inheritance

```python
class Material:
    def __init__(self, nama):
        self.nama = nama

    def info_material(self):
        print(f"Material: {self.nama}")

class Kayu(Material):
    def __init__(self, nama, jenis):
        super().__init__(nama)
        self.jenis = jenis

    def info_kayu(self):
        self.info_material()
        print(f"Jenis Kayu: {self.jenis}")

class MejaKayu(Kayu):
    def __init__(self, nama, jenis, ukuran):
        super().__init__(nama, jenis)
        self.ukuran = ukuran

    def info_meja_kayu(self):
        self.info_kayu()
        print(f"Ukuran Meja: {self.ukuran}")

# Membuat objek dari kelas MejaKayu
meja_saya = MejaKayu("Kayu Jati", "Jati", "120 x 60 cm")

# Memanggil metode untuk menampilkan informasi
meja_saya.info_meja_kayu()
```

# Implementasi Pewarisan Hierarchy Inheritance

```python
class Hewan:
    def __init__(self, nama):
        self.nama = nama

    def bersuara(self):
        print("Suara hewan...")

class Kucing(Hewan):
    def __init__(self, nama):
        super().__init__(nama)

    def bersuara(self):
        print("Meong...")

class Anjing(Hewan):
    def __init__(self, nama):
        super().__init__(nama)

    def bersuara(self):
        print("Gukguk...")

# Membuat objek dari kelas Kucing dan Anjing
kucing_saya = Kucing("Kitty")
anjing_saya = Anjing("Buddy")

# Memanggil metode bersuara()
kucing_saya.bersuara()  # Output: Meong...
anjing_saya.bersuara()  # Output: Gukguk...
```

# Membuat 2 Objek untuk menggunakan 2 Method dari 2 Class Turunan

```python
class Karyawan:
    def __init__(self, nama, id_karyawan):
        self.nama = nama
        self.id_karyawan = id_karyawan

    def tampilkan_info(self):
        print(f"Nama: {self.nama}, ID Karyawan: {self.id_karyawan}")

    def hitung_gaji(self):
        print("Menghitung gaji karyawan...")

class KaryawanTetap(Karyawan):
    def __init__(self, nama, id_karyawan, gaji_pokok):
        super().__init__(nama, id_karyawan)
        self.gaji_pokok = gaji_pokok

    def hitung_gaji(self):
        print(f"Gaji Karyawan Tetap {self.nama}: {self.gaji_pokok}")

    def tampilkan_status(self):
        print(f"{self.nama} adalah Karyawan Tetap")

class KaryawanKontrak(Karyawan):
    def __init__(self, nama, id_karyawan, upah_per_jam, jumlah_jam):
        super().__init__(nama, id_karyawan)
        self.upah_per_jam = upah_per_jam
        self.jumlah_jam = jumlah_jam

    def hitung_gaji(self):
        gaji = self.upah_per_jam * self.jumlah_jam
        print(f"Gaji Karyawan Kontrak {self.nama}: {gaji}")

    def tampilkan_status(self):
        print(f"{self.nama} adalah Karyawan Kontrak")


# Membuat 2 objek dari 2 kelas turunan
karyawan_tetap = KaryawanTetap("Andi", "kartap", 5000000)
karyawan_kontrak = KaryawanKontrak("Budi", "kontrak", 23000 ,100 )

# Menggunakan 2 metode dari objek karyawan_tetap
karyawan_tetap.tampilkan_info()
karyawan_tetap.hitung_gaji()
karyawan_tetap.tampilkan_status()


# Menggunakan 2 metode dari objek karyawan_kontrak
karyawan_kontrak.tampilkan_info()
karyawan_kontrak.hitung_gaji()
karyawan_kontrak.tampilkan_status()
```
